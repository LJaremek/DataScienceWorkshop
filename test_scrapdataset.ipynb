```python
from tools import ScrapDataset

image_files = {
    "gm": "gm.png",
    "osm": "osm.png",
    "cropped_gm": "cropped_gm.png",
    "cropped_osm": "cropped_osm.png",
    "cropped_osm_mask": "cropped_osm_mask.png"
}

dataset = ScrapDataset(image_files, "./data")

print("Len:", len(dataset))
```

```python
import math

import matplotlib.pyplot as plt
from PIL import Image
import numpy as np


def show_images(sample: dict[str, Image.Image]):
    """
    Wyświetla obrazy z przekazanego słownika w siatce z dwoma obrazkami w rzędzie.
    
    Args:
        sample (dict[str, Image.Image]): Słownik, w którym kluczami są nazwy obrazków,
            a wartościami obiekty PIL.Image.
    """
    n_images = len(sample)
    nrows = math.ceil(n_images / 2)

    _, axes = plt.subplots(nrows, 2, figsize=(10, 5 * nrows))
    
    if nrows == 1:
        axes = [axes[0], axes[1]]
    else:
        axes = axes.flatten()

    for ax, (name, img) in zip(axes, sample.items()):
        if img is not None:
            ax.imshow(np.array(img))
            ax.set_title(name)
        else:
            ax.text(0.5, 0.5, f'{name} not found', ha='center', va='center')
        ax.axis('off')
    
    for ax in axes[len(sample):]:
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
sample = dataset[0]
show_images(sample)
```
